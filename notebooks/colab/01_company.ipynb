## 회사 간 비교 (compare)

`c.panel` 이 한 회사를 항목×기간으로 수평화한다면, `dartlab.compare` 는 2~6개 회사를 같은 토픽·시점 격자로 정렬한다 — `scan` 처럼 단어 1개 톱레벨 verb. 결손은 NaN 유지(honest-gap).

```python
# row 모드 — 주석/서술 행을 disclosureKey 정렬키로 한 행 (라벨 drift 자동 해소)
dartlab.compare(["005930", "000660"], topic="재고")
```

```python
# 재무 셀 모드 — acode 단위·원 환산 (단위·라벨 착시 제거), scope 명시 고정
dartlab.compare(["005930", "000660"], topic="is", freq="year")
```

```python
# 다기간 — 셀 컬럼이 {code}␟{period} (회사 × 시점)
dartlab.compare(["005930", "000660"], topic="유형자산", period=["2025Q4", "2024Q4"])
```
