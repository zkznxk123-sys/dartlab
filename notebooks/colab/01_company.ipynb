```python
import dartlab
c = dartlab.Company("005930")
c.corpName
```

## 단일 진입점 — show / select

```python
# 사용 가능한 topic 목록
c.topics
```

```python
# 손익계산서 (분기)
c.show("IS")
```

```python
# 연간 합산
c.show("IS", freq="Y")
```

```python
# 행 필터
c.select("IS", ["매출액", "영업이익", "당기순이익"])
```

```python
# 주석 — 재고자산
c.show("inventory")
```

```python
# 정형 공시 — 배당
c.show("dividend")
```
