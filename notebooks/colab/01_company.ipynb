```python
import dartlab
c = dartlab.Company("005930")
c.corpName
```

## 격자 + 행 검색

```python
# 전체 공시 수평화 격자 (항목 × 기간)
c.panel()
```

```python
# 항목명 행 검색 (raw 공시)
c.panel("매출")
```

## native 재무제표 (소문자) / finance (대문자)

```python
# native 손익 — 사업보고서 항목 그대로, XBRL+옛 통합 (2013~)
c.panel("is", freq="year")
```

```python
# finance 손익 — XBRL 정규화 숫자
c.panel("IS", freq="year")
```

## native 재무비율 (소문자) / finance 비율 (대문자)

```python
# native 비율 — 5표 항목으로 계산 (깊은 history)
c.panel("ratios")
```

```python
# finance 비율
c.panel("RATIOS")
```
