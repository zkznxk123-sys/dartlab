## 수급 (KR 전용)

```python
# 최근 수급 (기본 5거래일)
dartlab.gather("flow", "005930")
```

```python
# 2010년부터 최신 거래일까지 — 자동 페이지네이션
dartlab.gather(
    "flow",
    "005930",
    start="2010-01-04",
    sleepSec=1.0,
)
```
